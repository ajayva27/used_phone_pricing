```python
x=data.drop('normalized_used_price',axis=1)
y=data[['normalized_used_price']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
x['device_brand']=label.fit_transform(x['device_brand'])
x['os']=label.fit_transform(x['os'])
x.head()
```

```python
# x=pd.get_dummies(x,columns=['device_brand','os'],drop_first=True,dtype='int32')
# x.head()
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
linear_model=LinearRegression()
linear_model.fit(xtrain,ytrain)
```

```python
lr_train_pred=linear_model.predict(xtrain)
lr_test_pred=linear_model.predict(xtest)
lr_train_score=linear_model.score(xtrain,ytrain)
lr_test_score=linear_model.score(xtest,ytest)
lr_mse=mean_squared_error(ytest,lr_test_pred)
lr_r2_score=r2_score(ytest,lr_test_pred)
```

```python
print("train score:",lr_train_score)
print("test score:",lr_test_score)
print("MSE:",lr_mse)
print("R2 score",lr_r2_score)
```

```python
data_c=ytest.copy()
data_c["pred"]=lr_test_pred
data_c['residual']=data_c['normalized_used_price']-data_c["pred"]
data_c.head()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_c['pred'],data_c['residual'],color='red')
plt.title("Predicted vs Residual")
plt.xlabel("predicted")
plt.ylabel("residual")
plt.grid(True)
plt.show()
```
